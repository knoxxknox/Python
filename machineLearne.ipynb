```python
# Carregando os dados do arquivo CSV.
# https://www.kaggle.com/dell4010/wine-dataset

import pandas as pd
arquivo = pd.read_csv('/Users/knox/wine_dataset.csv')
```

```python
arquivo.head()
```

```python
# Subistituindo os valores string por numericos na coluna style

arquivo['style'] = arquivo['style'].replace('red',0)
arquivo['style'] = arquivo['style'].replace('white',1)
```

```python
# Verificando o arquivo após a substituiçao na coluna style
arquivo.head()
```

```python
# Separando as variáveis em preditoras e alvo, onde alvo é a variável que se quer prever, neste caso style

y = arquivo['style']
x = arquivo.drop('style', axis = 1)
```

```python
# separando 30% do conjunto de dados para testes, e 70% para treino
# a função train_test_split já nos da 4 saida

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Criação do modelo, para rodar no algorítimo de machine learne "ExtraTreesClassifier"
# função fit utilizada para aplicar o algotítmo "ExtraTreesClassifier" aos dados de treino previamente selecionados

modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino,y_treino)

resultado = modelo.score(x_teste, y_teste)
print("Taxa de acerto: ", resultado)


```
