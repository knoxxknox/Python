```python
# https://www.kaggle.com/pavanraj159/predicting-a-pulsar-star/version/1

import pandas as pd

arquivo = pd.read_csv('/Users/knox/pulsar_stars.csv')
```

```python
# target_class9.class of pulsar star. 1 for pulsar star,0 for not a star

arquivo.head()
```

```python
arquivo.shape
```

```python
# Separando as variáveis em preditoras e alvo, onde alvo é a variável que se quer prever, neste caso style

x = arquivo.drop('target_class', axis = 1)
y = arquivo['target_class']
```

```python
# separando 20% do conjunto de dados para testes, e 80% para treino
# a função train_test_split já nos da 4 saida

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.9)
```

```python
x_treino.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Criação do modelo, para rodar no algorítimo de machine learne "ExtraTreesClassifier"
# função fit utilizada para aplicar o algotítmo "ExtraTreesClassifier" aos dados de treino previamente selecionados

modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino,y_treino)

resultado = modelo.score(x_teste, y_teste)
print("Taxa de acerto: ", resultado)
```
